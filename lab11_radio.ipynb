### Step 2 — Clean the Data

U.S. broadcast callsigns all begin with **K** or **W**.  
Any rows where the callsign starts with **D** are non-conforming entries that should be dropped before analysis.

````python
# Check how many non-conforming rows exist before cleaning
bad_rows = stations[stations['callsign'].str.startswith('D', na=False)]
print(f"Non-conforming rows (callsign starts with D): {len(bad_rows)}")
bad_rows[['callsign', 'band', 'city']].head()
````

````python
# Drop those rows and reset the index
stations = stations[~stations['callsign'].str.startswith('D', na=False)].reset_index(drop=True)

print(f"Stations after cleaning: {len(stations):,}")
````

### Step 3 — Fix Longitude

In the dataset, longitude values are stored as **positive** numbers.  
In the Western Hemisphere (where the U.S. is), longitude is actually **negative**.  
We need to negate the `lon` column so geographic filtering works correctly later.

````python
print("Before fix — sample lon values:", stations['lon'].head(3).tolist())

stations['lon'] = stations['lon'] * -1

print("After fix  — sample lon values:", stations['lon'].head(3).tolist())
````
