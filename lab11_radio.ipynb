### A Note on `and` / `or` in Python

In plain Python, you combine conditions in an `if` statement using `and` and `or` — just like the words sound:

```python
band = 'FM'
power = 75

if band == 'FM' and power > 50:
    print("High-power FM station")

if band == 'AM' or power < 1:
    print("Low-reach station")
```

**So why does Pandas use `&` and `|` instead of `and`/`or`?**  
When you filter a DataFrame, each condition produces an entire column of True/False values — one per row — not a single True or False. Python's `and`/`or` can only compare two single booleans. The `&` and `|` operators apply the logic row-by-row across the whole column at once.

````python
# Plain Python and / or — no DataFrame needed
band = 'FM'
power = 75

if band == 'FM' and power > 50:
    print("High-power FM station")

if band == 'AM' or power < 1:
    print("Low-reach station")
else:
    print("Neither condition was true")
````
